# Trying Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Function to pull book info
def get_book_info(book_id):
    url = f'https://www.google.com/books/edition/_/{book_id}?hl=en'

    # Creates a new Chrome webdriver
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    driver = webdriver.Chrome(options=options)

    # Pulls data
    driver.get(url)

    # Clicks on the "More" button to expand the subject field
    try:
        more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'span[data-t="kno-fv-exp"]')))
        more_button.click()
    except:
        pass

    # Extracts the book information
    info_elements = driver.find_elements(By.CSS_SELECTOR, 'span.LrzXr.kno-fv.wHYlTd.z8gr9e')
    if len(info_elements) >= 1:
        isbn = None
        page_count = None
        publishing_date = None
        form = None
        publisher = None
        language = None
        author = None
        illustrator = None
        originally_published = None
        genres = None
        subject = None
        counter = 0
        for i in info_elements:
            counter += 1
            if counter == 1:
                isbn = i.text.strip()
            elif counter == 2:
                page_count = i.text.strip()
            elif counter == 3:
                publishing_date = i.text.strip()
            elif counter == 4:
                form = i.text.strip()
            elif counter == 5:
                publisher = i.text.strip()
            elif counter == 6:
                language = i.text.strip()
            elif counter == 7:
                author = i.text.strip()
            elif counter == 8:
                illustrator = i.text.strip()
            elif counter == 9:
                originally_published = i.text.strip()
            elif counter == 10:
                genres = i.text.strip()
            elif counter == 11:
                subject = i.text.strip()

        # Pulls description
        description = driver.find_elements(By.CSS_SELECTOR, 'div.Y0Qrof')

        # Creates a dictionary of the book information
        book_info = {
            'id': book_id,
            'isbn': isbn,
            'page_count': page_count,
            'publishing_date': publishing_date,
            'form': form,
            'publisher': publisher,
            'language': language,
            'author': author,
            'illustrator': illustrator,
            'originally_published': originally_published,
            'genres': genres,
            'subject': subject,
            'description': description[0].text.strip()
        }

        # Returns dictionary
        return book_info
    else:
        # Prints error
        print("Book information not found")
        return None
```

```python
# Define a list of book IDs
book_ids = ["dyevo2Bz6a4C", "Iw_gHtk4ghYC", "ayJpGQeyxgkC"]
```

```python
# Create an empty list to store the book_info dictionaries
book_info_list = []

# Loop over the book_ids and call the get_book_info function for each id
for book_id in book_ids:
    book_info = get_book_info(book_id)
    if book_info is not None:
        book_info_list.append(book_info)
```

```python
# Create a dataframe from the list of dictionaries
df = pd.DataFrame(book_info_list)
```
