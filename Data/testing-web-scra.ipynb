```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from requests_html import HTMLSession
from selectolax.parser import HTMLParser
```

```python
#Grabbing one book
book_id = "dyevo2Bz6a4C"

#url_base
url = f'https://www.google.com/books/edition/_/{book_id}?hl=en'

#pulling data
res = requests.get(url)

#Checking response
res.text
```

```python
#Some soup!!!!
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
for books in soup:
    print(soup.select_one("h1 class").attrs["booktitle"])
```

```python
#pulling title
title = soup.find('h1', {'class': 'booktitle'}).text.strip()

#adding author condition
author_tag = soup.find('a', {'class': 'fl'})
author = author_tag.text.strip() if author_tag else ''

#adding description condition
description_tag = soup.find('div', {'class': 'description'})
description = description_tag.text.strip() if description_tag else ''

data = {'title': title, 'author': author, 'description': description}
data
```
