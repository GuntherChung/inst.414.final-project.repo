```python
import pandas as pd
```

```python
df = pd.read_csv('/Data/isbn13_results.csv')
df.head()
```

```python
descriptions = list(df['description'])
empty = 0
full = 0
for i in descriptions:
    #print(i)
    if pd.isna(i):
        empty+=1
    else:
        full += 1
        
print(empty, full)
```

```python
pages = list(df['page_count'])
empty = 0
full = 0
for i in pages:
    #print(i)
    if pd.isna(i):
        empty+=1
    else:
        full += 1
        
print(empty, full)
```

```python
ratings = list(df['average_rating'])
empty = 0
full = 0
for i in ratings:
    #print(i)
    if pd.isna(i):
        empty+=1
    else:
        full += 1
        
print(empty, full)
```

```python
# Remove rows with NaNs
descrNan = df.loc[pd.isna(df['description'])]
descrNan.head()
```

```python
pagesNan = df.loc[pd.isna(df['page_count'])]
pagesNan.head()
```

```python
ratingsNan = df.loc[pd.isna(df['ratings_count'])]
ratingsNan.head()
```

```python
indexNan = []
for i in descrNan.index:
    indexNan.append(i)
for i in pagesNan.index:
    indexNan.append(i)
for i in ratingsNan.index:
    indexNan.append(i)
```

```python
indexNan = [*set(indexNan)]
```

```python
indexNan.sort()
```

```python
indexNan
```

```python
for i in indexNan:
    df = df.drop([i])
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
df.drop(columns=['index', 'Unnamed: 0'], inplace=True)
df.head()
```
